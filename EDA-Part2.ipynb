## Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
data=pd.read_csv('Iris.csv',index_col=0)
data_encoded=pd.get_dummies(data) #converting categorical text data to categorical numeric data
```

```python
#training the model
clf=IsolationForest(random_state=10,contamination=0.01)
clf.fit(data_encoded)
```

```python
#predictions
y_pred_outliers=clf.predict(data_encoded)
```

```python
#-1 for outliers and 1 for inliers
y_pred_outliers
```

#### In the above o/p, there are 2 values which has -1 means there are 2 outliers which has been identified

```python
data_encoded
```

```python
#Adding new data points to the 150th record as the outlier
data_encoded.loc[150]=[20,40,30,50,1,0,0]
```

```python
data_encoded
```

```python
#rebuilding the model with new data_encoded
clf=IsolationForest(random_state=10,contamination=0.01)
clf.fit(data_encoded)
#predictions
y_pred_outliers=clf.predict(data_encoded)
y_pred_outliers
```

#### Thus, in the above o/p last(150th) record has been identified as the outlier

```python
#calculating the anomaly score
data_encoded['scores']=clf.decision_function(data_encoded)
```

```python
data_encoded['anomaly']=clf.predict(data_encoded.iloc[:,0:7])
```

```python
data_encoded
```

#### In the above o/p, last record(150th) has an anomaly of -1, which indicates that it is an outlier

```python
#Print the outlier data points
data_encoded[data_encoded['anomaly']==-1]
```
